# Exploratory Data Analysis on Global Renewable Energy and Indicators Dataset

## 1. Introduction

### 1.1 Dataset Description

The Global Renewable Energy and Indicators dataset, obtained from Kaggle, contains comprehensive information on various aspects of renewable energy production and its related indicators globally. The dataset includes data on energy production, installed capacity, economic indicators, environmental factors, and much more. It spans multiple years and covers numerous countries and regions, providing a detailed view of the global renewable energy landscape.

#### Key Columns
- **Country**: The country for which the data is reported.
- **Year**: The year of the data entry.
- **Energy Type**: The type of energy (e.g., solar, wind, hydro).
- **Production (GWh)**: The amount of energy produced (in gigawatt hours).
- **Installed Capacity (MW)**: The installed capacity of renewable energy sources (in megawatts).
- **Investments (USD)**: Investments made in renewable energy (in USD).
- **Population**: The population of the country.
- **GDP**: Gross Domestic Product of the country (in USD).
- **Energy Consumption**: Total energy consumption.
- **Energy Exports**: Energy exported by the country.
- **Energy Imports**: Energy imported by the country.
- **CO2 Emissions**: CO2 emissions from the country (in metric tons).
- **Renewable Energy Jobs**: Number of jobs in the renewable energy sector.
- **Government Policies**: Policies promoting renewable energy.
- **R&D Expenditure**: Expenditure on research and development in renewables.
- **Renewable Energy Targets**: National targets for renewable energy adoption.
- **Environmental and Climatic Factors**: Including average annual temperature, annual rainfall, solar irradiance, wind speed, hydro potential, geothermal potential, biomass availability.
- **Market and Policy Indicators**: Electricity prices, energy subsidies, international aid, public awareness, energy efficiency programs.
- **Socio-Economic and Institutional Factors**: Urbanization rate, industrialization rate, market liberalization, patents, educational levels, technology transfer agreements, local manufacturing capacity, tariffs, incentives, political stability, corruption perception, regulatory quality, rule of law, economic freedom, ease of doing business, innovation index, number of research institutions, and more.

### 1.2 Objectives of the Analysis

The primary objective of this exploratory data analysis (EDA) is to understand the distribution, relationships, and patterns within the dataset. Specifically, the analysis aims to:

- Identify trends in renewable energy production, installed capacity, and investments over time and across different regions.
- Investigate the relationship between renewable energy production and economic indicators such as GDP, investments, and energy consumption.
- Examine the influence of environmental factors (e.g., solar irradiance, wind speed) on renewable energy production.
- Explore the correlation between renewable energy policies, R&D expenditure, and renewable energy production capacity.
- Analyze the impact of socio-economic and institutional factors on renewable energy adoption.
- Identify any outliers or anomalies in the data.
- Provide actionable insights or recommendations based on the findings.

### 1.3 Key Questions

During this analysis, we aim to answer the following key questions:

1. **Trends Over Time**:
   - How has renewable energy production and installed capacity changed over the years?
   - What trends are observed in renewable energy investments over time?

2. **Economic Impact**:
   - Is there a correlation between GDP and renewable energy production?
   - How do investments in renewable energy impact overall energy production and installed capacity?

3. **Regional Differences**:
   - Which countries and regions are leading in renewable energy production and investments?
   - Are there significant differences in renewable energy production between different regions and energy types?

4. **Environmental Influence**:
   - How do environmental and climatic factors like solar irradiance and wind speed impact renewable energy production?

5. **Policy and Socio-Economic Factors**:
   - What is the impact of government policies, R&D expenditure, and renewable energy targets on production capacity?
   - How do socio-economic factors like urbanization rate, political stability, and regulatory quality influence renewable energy adoption?

6. **Population and Emissions**:
   - How does population impact renewable energy production and CO2 emissions?
   - What is the relationship between renewable energy production and CO2 emissions?

7. **Market and Incentives**:
   - How do electricity prices, energy subsidies, and international aid affect renewable energy investments and production?

By conducting this exploratory data analysis, we hope to gain valuable insights into the global landscape of renewable energy production and its interactions with economic, environmental, policy, and socio-economic factors.

## 3. Data Cleaning

### 3.1 Handling Missing Values

Checking for Missing Values

```python
# checking for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values[missing_values > 0])
```

* There are no missing values

### 3.2 Handling Outliers

Handling Outliers Using Z-scores

```python
from scipy import stats
# Calculate Z-scores
z_scores = np.abs(stats.zscore(df[numerical_columns]))

# Filter out rows where any column has a Z-score greater than 3
data_no_outliers = df[(z_scores < 3).all(axis=1)]
print(f"Data shape after removing outliers using Z-scores: {data_no_outliers.shape}")
```

* There are no outliers

## 5. Bivariate and Multivariate Analysis

### 5.1 Relationships Between Two Variables

#### 5.1.2 Scatter Plots

```python
### uncomment if you want to see the current correlation - NOT RECOMENDED
# corr = df.corr()
# corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
# Group by 'Country' and 'Year' to have a better view of the correlation between the variables
grouped_data = df.groupby(['Country', 'Year']).agg({
    'Production (GWh)': 'sum',           # Sum of energy production per country per year
    'Installed Capacity (MW)': 'sum',    # Sum of installed capacity
    'Investments (USD)': 'sum',          # Total investments
    'CO2 Emissions': 'sum',               # Sum of CO2 emissions
    'Population': 'mean',                 # Average population (or 'sum' if preferred)
    'GDP': 'mean',                        # Average GDP (or 'sum' if preferred)
    'Energy Consumption': 'sum'           # Sum of energy consumption per country per year
}).reset_index()

corr = grouped_data.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
# Scatter plots for key variable pairs
# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # 2 rows, 2 columns



# Scatter for Investments vs Installed Capacity
sns.scatterplot(x='Investments (USD)', y='Installed Capacity (MW)', data=grouped_data, ax=axs[0, 0], alpha=0.6, s=50)
axs[0,0].set_title('Investments (USD) vs Installed Capacity (MW)')
axs[0,0].set_xlabel('Investments (USD)')
axs[0,0].set_ylabel('Installed Capacity (MW)')


#Scatter for Production and Investments
sns.scatterplot(x='Production (GWh)', y='Investments (USD)', data=grouped_data, ax=axs[0,1], alpha=0.6, s=50)
axs[0,1].set_title('Production (GWh) vs Investments (USD)')
axs[0,1].set_xlabel('Production (GWh)')
axs[0,1].set_ylabel('Investments (USD)')


#Scatter for Population vs Energy Consumption
sns.scatterplot(x='Population', y='Energy Consumption', data=grouped_data, ax=axs[1,0], alpha=0.6, s=50)
axs[1,0].set_title('Population vs Energy Consumption')
axs[1,0].set_xlabel('Population')
axs[1,0].set_ylabel('Energy Consumption')


#Scatter for CO2 Emissions vs Production
sns.scatterplot(x='CO2 Emissions', y='Production (GWh)', data=grouped_data, ax=axs[1,1], alpha=0.6, s=50)
axs[1,1].set_title('CO2 Emissions vs Production (GWh)')
axs[1,1].set_xlabel('CO2 Emissions')
axs[1,1].set_ylabel('Production (GWh)')

plt.tight_layout()
plt.show()
```
